```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
rs['keywords']=rs['keywords'].apply(convert)
rs['genres']=rs['genres'].apply(convert)

rs
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i ['job'] == "Director":
            L.append(i['name'])
    return L
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def more_clean(list):
    list = [words for words in list if not words in stop_words]
    list = [stemmer.stem(words) for words in list]
    return list
```

```python
rs['genres'] = rs['genres'].apply(more_clean)
rs['keywords'] = rs['keywords'].apply(more_clean)
rs['overview'] = rs['overview'].apply(more_clean)


rs
```

```python
def get_film_vector(idx, ser, model):
    vector = np.zeros(100)
    word_list = ser.iloc[idx].split()
    for words in word_list:
        vector += model.wv.get_vector(words)

    return vector
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models.word2vec import Word2Vec
from scipy import spatial
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
def get_recommendations(title, weights=[1, 1, 1, 1, 1], model=1):
    Ser = weights[0] * rs['genres'] + weights[1] * rs['keywords'] + \
          weights[2] * rs['overview'] + weights[3] * rs['cast'] + \
          weights[4] * rs['director']

    Ser_lda = weights[2] * rs['overview']



    indices = pd.Series(rs.index, index=rs['title']).drop_duplicates()


    if(model == 1):


        tfidf = TfidfVectorizer()
        tfidf_matrix = tfidf.fit_transform(Ser)
        cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

        idx = indices[title]

        sim_scores = enumerate(cosine_sim[idx])
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:11]
        movie_indices = [i[0] for i in sim_scores]

        for vals in rs['title'].iloc[movie_indices].values:
            print(vals)

    elif model == 2:


        data = []
        for summs in Ser.values:
            lst = summs.split()
            data.append(lst)

        vec_model = Word2Vec(data, window=10, min_count=1, workers=4, sg=1)

        # Check if the title is present in the DataFrame
        if title in indices:
            idx = indices.loc[title]
            # Check if the index is within the valid range
            if 0 <= idx < len(rs):
                the_vec = get_film_vector(idx, Ser, vec_model)
                sims = []
                for i in range(len(rs)):
                    vec2 = get_film_vector(i, Ser, vec_model)
                    cossim = 1 - spatial.distance.cosine(the_vec, vec2)
                    sims.append((cossim, i))

                sims = sorted(sims)
                recclist = []
                for cosi, ind in sims[-11:-1]:
                    recclist.append(rs['title'].values[ind])

                recclist.reverse()

                for rex in recclist:
                    print(rex)
            else:
                print(f"Index '{idx}' is out of bounds. Number of rows in DataFrame: {len(rs)}")
        else:
            print(f"Title '{title}' not found in the DataFrame. Titles in DataFrame: {list(rs['title'])}")


    elif model == 3:



        lda_vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
        lda_dtm = lda_vectorizer.fit_transform(Ser_lda)

        lda_model = LatentDirichletAllocation(n_components=5, random_state=42)
        lda_model.fit(lda_dtm)




        def get_film_vector_lda(overview, vectorizer, model):
            lda_vector = model.transform(vectorizer.transform([overview]))[0]
            return lda_vector




        def get_top_movies_lda(lda_vector, n=10):
            # Assuming 'lda_vector' is a column in the DataFrame to store LDA vectors
            rs['lda_vector'] = rs.apply(lambda row: get_film_vector_lda(row['overview'], lda_vectorizer, lda_model), axis=1)


            cosine_sim_lda = cosine_similarity([lda_vector], list(rs['lda_vector']))[0]

            # Get indices of top similar movies
            top_movie_indices_lda = sorted(range(len(cosine_sim_lda)), key=lambda i: cosine_sim_lda[i], reverse=True)[1:n+1]

            # Get the titles of the top movies
            top_movies_lda = rs['title'].iloc[top_movie_indices_lda].values

            return top_movies_lda

        # Check if the title is present in the DataFrame
        if title in indices:
            idx = indices.loc[title]
            # Check if the index is within the valid range
            if 0 <= idx < len(rs):
                lda_vector = get_film_vector_lda(rs['overview'].iloc[idx], lda_vectorizer, lda_model)

                # Get top movies based on LDA vector
                top_movies_lda = get_top_movies_lda(lda_vector, n=10)

                for movie in top_movies_lda:
                    print(movie)
            else:
                print(f"Index '{idx}' is out of bounds. Number of rows in DataFrame: {len(rs)}")
        else:
            print(f"Title '{title}' not found in the DataFrame. Titles in DataFrame: {list(rs['title'])}")








```
